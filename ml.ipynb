```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, precision_score, recall_score, roc_auc_score, roc_curve, accuracy_score

# 1. Load dataset
df = pd.read_csv("data.csv")

# Drop unnamed column if present
df = df.drop(columns=["Unnamed: 32"], errors="ignore")

# Convert diagnosis to numeric (M=1, B=0)
df["diagnosis"] = df["diagnosis"].map({"M":1, "B":0})

X = df.drop(columns=["id","diagnosis"])
y = df["diagnosis"]

# 2. Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)

# 3. Standardize
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)

# 4. Logistic Regression
model = LogisticRegression(solver='liblinear', random_state=42)
model.fit(X_train_s, y_train)

# 5. Predictions
y_proba = model.predict_proba(X_test_s)[:, 1]
y_pred = (y_proba >= 0.5).astype(int)

# 6. Metrics
cm = confusion_matrix(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_proba)

print("Confusion Matrix:\n", cm)
print(f"Accuracy: {acc:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"ROC-AUC: {roc_auc:.4f}")

# 7. ROC curve
fpr, tpr, _ = roc_curve(y_test, y_proba)
plt.plot(fpr, tpr, label=f"ROC curve (AUC={roc_auc:.2f})")
plt.plot([0,1],[0,1],'--',color='gray')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()

# 8. Threshold tuning
thresholds = [0.9,0.7,0.5,0.3]
for thr in thresholds:
    y_pred_thr = (y_proba >= thr).astype(int)
    cm_thr = confusion_matrix(y_test, y_pred_thr)
    prec_thr = precision_score(y_test, y_pred_thr)
    rec_thr = recall_score(y_test, y_pred_thr)
    print(f"\nThreshold={thr}:")
    print("Confusion Matrix:\n", cm_thr)
    print(f"Precision={prec_thr:.4f}, Recall={rec_thr:.4f}")
```
